## Matplotlib

This is probably the most widely used plotting toolkit. It is what is loaded when we start a notebook with the `--pylab=inline` argument. Matplotllib has two interfaces:
* a functional API that mimics Matlab (pylab)
* an object-oriented version (matplotlib.pyplot, typically aliased to plt)

Several examples are adapted from <http://matplotlib.org/gallery.html>

### Line graph

```python
try:
    nile = pd.read_pickle('nile.pic')
except:
    nile = sm.datasets.get_rdataset('Nile')
    nile = pd.DataFrame(nile.data)
    nile.to_pickle('nile.pic')
    
window = 20
offset = nile.time[0] + 20
ma = pd.rolling_mean(nile.Nile, window)
mstd = pd.rolling_std(nile.Nile, window)
plt.plot(nile.time, nile.Nile, 'k')
plt.plot(offset+ma.index, ma, 'b')
plt.fill_between(offset+mstd.index, ma-2*mstd, ma+2*mstd, color='b', alpha=0.2)
xlabel('Year', fontsize=16)
ylabel('Flow', fontsize=16)
title('Annual flow of river Nile at Aswan', fontsize=16);
```

### Adding interactivity

A brushing example - Mouse over a region - the brushed area will be highlighted in all panels simultaneously. Also, notice that a toolbar with tools for panninng, zooming, brushing and resetting (home icon) appear as you hover over the image.

Example from <http://mpld3.github.io/notebooks/mpld3_demo.html>

```python
import mpld3
from mpld3 import plugins

fig, ax = plt.subplots(3, 3, figsize=(6, 6))
fig.subplots_adjust(hspace=0.1, wspace=0.1)
ax = ax[::-1]

X = np.random.normal(size=(3, 100))
for i in range(3):
    for j in range(3):
        ax[i, j].xaxis.set_major_formatter(plt.NullFormatter())
        ax[i, j].yaxis.set_major_formatter(plt.NullFormatter())
        points = ax[i, j].scatter(X[j], X[i])
        
plugins.connect(fig, plugins.LinkedBrush(points))
mpld3.display(fig)
```

### Basemap

Example below is from <https://github.com/koldunovn/python_for_geosciences>

```python
datafile = 'air.sig995.2012.nc'
if not os.path.exists(datafile):
    !wget ftp://ftp.cdc.noaa.gov/Datasets/ncep.reanalysis/surface/air.sig995.2012.nc
f = netcdf.netcdf_file(datafile)
air = f.variables['air']
lat = f.variables['lat'][:]
lon = f.variables['lon'][:]
lon2, lat2 = np.meshgrid(lon,lat)
air_c = ((air[:] * air.scale_factor) + air.add_offset) - 273.15

plt.jet()
fig = plt.figure(figsize=(15,7))
m = Basemap(projection='ortho',lat_0=45,lon_0=-100,resolution='l')
x, y = m(lon2, lat2)
m.fillcontinents(color='lightgrey',lake_color='white')
m.drawcoastlines()
m.drawparallels(np.arange(-80.,81.,20.))
m.drawmeridians(np.arange(-180.,181.,20.))
m.drawmapboundary(fill_color='white')
cs = m.contourf(x,y,air_c[0,:,:],20)
plt.colorbar()
plt.title('Monthly mean SAT');
```

### 3D plots

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3, cmap=cm.jet)
cset = ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100);
```

## Seaborn

This new library is designed for statistical graphics. Note that `import seaborn` resets Matplotlib defaults to be more "R"-like. To get back Matplotlib style plots, call `matplotlib.rcdefaults() `

### Plot combining KDE with rug/histogram

```python
import seaborn as sns

random.seed(23)
xs = random.normal(0,1,100)

figure(figsize=(8,4))
sns.set_palette("hls")
plt.subplot(1,2,1)
sns.distplot(xs, hist=False, rug=True);
sns.set_palette("coolwarm")
plt.subplot(1,2,2)
sns.distplot(xs, hist=True);
```
