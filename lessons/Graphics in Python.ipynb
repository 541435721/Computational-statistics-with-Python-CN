## Matplotlib

This is probably the most widely used plotting toolkit. It is what is loaded when we start a notebook with the `--pylab=inline` argument. Matplotllib has two interfaces:
* a functional API that mimics Matlab (pylab)
* an object-oriented version (matplotlib.pyplot, typically aliased to plt)

Several examples are adapted from <http://matplotlib.org/gallery.html>

### Line graph

````python
try:
    nile = pd.read_pickle('nile.pic')
except:
    nile = sm.datasets.get_rdataset('Nile')
    nile = pd.DataFrame(nile.data)
    nile.to_pickle('nile.pic')
    
window = 20
offset = nile.time[0] + 20
ma = pd.rolling_mean(nile.Nile, window)
mstd = pd.rolling_std(nile.Nile, window)
plt.plot(nile.time, nile.Nile, 'k')
plt.plot(offset+ma.index, ma, 'b')
plt.fill_between(offset+mstd.index, ma-2*mstd, ma+2*mstd, color='b', alpha=0.2)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Flow', fontsize=16)
plt.title('Annual flow of river Nile at Aswan', fontsize=16);
````

### Bar chart

````python
df = sm.datasets.longley.load_pandas().data

c = plt.cm.get_cmap("jet", len(df))
cvals = c(np.arange(len(df)))
plt.bar(df['YEAR'], df['UNEMP'], align='center', alpha=0.5, color=cvals)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Number of unemployed ', fontsize=16)
plt.xlim([df['YEAR'].min()-1, df['YEAR'].max()+1]);
````

````python
# Scatter plot

df = sm.datasets.longley.load_pandas().data
plt.scatter(df['YEAR'], df['UNEMP'], alpha=0.5, 
            s=200.0*df['ARMED']/max(df['ARMED']),
            c=df['GNP'], cmap=plt.cm.jet)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Number of unemployed ', fontsize=16)
plt.xlim([df['YEAR'].min()-1, df['YEAR'].max()+1]);
````

### Contour plot

````python
x = np.linspace(-3.0, 3.0, 100) 
y = np.linspace(-3.0, 3.0, 100) 
X, Y = np.meshgrid(x, y) 
Z = np.exp(-(X**2 + Y**2)) + np.random.normal(0, 0.1, size=X.shape)
plt.contour(X, Y, Z)
plt.contourf(X, Y, Z, cmap=plt.cm.coolwarm)
plt.colorbar();
````

### Data interpolation for scatter plot

Example adapted from <http://matplotlib.org/examples/pylab_examples/tricontour_vs_griddata.html>.

````python
npts = 200
x = np.random.uniform(-2,2,npts)
y = np.random.uniform(-2,2,npts)
z = x*np.exp(-x**2-y**2)
xi = np.linspace(-2.1,2.1,100)
yi = np.linspace(-2.1,2.1,100)
# grid the data.
zi = plt.mlab.griddata(x, y, z, xi, yi)
# contour the gridded data, plotting dots at the randomly spaced data points.
CS = plt.contour(xi,yi,zi,15,linewidths=0.5,colors='k')
CS = plt.contourf(xi,yi,zi,15,cmap=plt.cm.jet)
plt.colorbar() # draw colorbar
plt.scatter(x,y,marker='o',c='b',s=5)
plt.xlim(-2,2)
plt.ylim(-2,2);
````

### Adding interactivity

A brushing example - Mouse over a region - the brushed area will be highlighted in all panels simultaneously. Also, notice that a toolbar with tools for panninng, zooming, brushing and resetting (home icon) appear as you hover over the image.

Example from <http://mpld3.github.io/notebooks/mpld3_demo.html>

````python
import mpld3
from mpld3 import plugins

fig, ax = plt.subplots(3, 3, figsize=(6, 6))
fig.subplots_adjust(hspace=0.1, wspace=0.1)
ax = ax[::-1]

X = np.random.normal(size=(3, 100))
for i in range(3):
    for j in range(3):
        ax[i, j].xaxis.set_major_formatter(plt.NullFormatter())
        ax[i, j].yaxis.set_major_formatter(plt.NullFormatter())
        points = ax[i, j].scatter(X[j], X[i])
        
plugins.connect(fig, plugins.LinkedBrush(points))
mpld3.display(fig)
````

### Basemap

Example below is from <https://github.com/koldunovn/python_for_geosciences>

````python
datafile = 'air.sig995.2012.nc'
if not os.path.exists(datafile):
    !wget ftp://ftp.cdc.noaa.gov/Datasets/ncep.reanalysis/surface/air.sig995.2012.nc
f = netcdf.netcdf_file(datafile)
air = f.variables['air']
lat = f.variables['lat'][:]
lon = f.variables['lon'][:]
lon2, lat2 = np.meshgrid(lon,lat)
air_c = ((air[:] * air.scale_factor) + air.add_offset) - 273.15

plt.jet()
fig = plt.figure(figsize=(15,7))
m = Basemap(projection='ortho',lat_0=45,lon_0=-100,resolution='l')
x, y = m(lon2, lat2)
m.fillcontinents(color='lightgrey',lake_color='white')
m.drawcoastlines()
m.drawparallels(np.arange(-80.,81.,20.))
m.drawmeridians(np.arange(-180.,181.,20.))
m.drawmapboundary(fill_color='white')
cs = m.contourf(x,y,air_c[0,:,:],20)
plt.colorbar()
plt.title('Monthly mean SAT');
````

### 3D plots

Example adapted from <http://matplotlib.org/examples/mplot3d/contourf3d_demo2.html>

````python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3, cmap=plt.cm.jet)
cset = ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap=plt.cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap=plt.cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=40, cmap=plt.cm.coolwarm)

ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100);
````

## Seaborn

This new library is designed for statistical graphics. Note that `import seaborn` resets Matplotlib defaults to be more "R"-like. To get back Matplotlib style plots, call `matplotlib.rcdefaults() `

### Plot combining KDE with rug/histogram

````python
import seaborn as sns

np.random.seed(23)
xs = np.random.normal(0,1,100)

plt.figure(figsize=(8,4))
sns.set_palette("hls")
plt.subplot(1,2,1)
sns.distplot(xs, hist=False, rug=True);
sns.set_palette("coolwarm")
plt.subplot(1,2,2)
sns.distplot(xs, hist=True);
````

### Trellis plots

````python
try:
    tips = pd.read_pickle('tips.pic')
except:
    tips = pd.read_csv('https://raw.github.com/vincentarelbundock/Rdatasets/master/csv/reshape2/tips.csv')
    tips.to_pickle('tips.pic')

plot = rplot.RPlot(tips, x='total_bill', y='tip')
plot.add(rplot.TrellisGrid(['sex', 'smoker']))
plot.add(rplot.GeomScatter())
plot.add(rplot.GeomPolyFit(degree=2))
plot.render(plt.gcf());
````

### Using ggplot2 from Python

If you are already familiar with ggplot2, you can use it with the ggplot package. Example from <https://github.com/yhat/ggplot>.

````python
from ggplot import meat, ggplot, aes, geom_point, stat_smooth

meat_lng = pd.melt(meat[['date', 'beef', 'pork', 'broilers']], id_vars='date')
ggplot(aes(x='date', y='value', colour='variable'), 
       data=meat_lng) + geom_point() + stat_smooth(color='red')
````

### Bokeh

Example is from <https://github.com/ContinuumIO/bokeh/blob/0.4.4/examples/ggplot/density.py>

See [tutorial](http://bokeh.pydata.org/tutorial/index.html) and [gallery](http://bokeh.pydata.org/docs/gallery.html).

````python
from ggplot import *
from bokeh import pyplot
import matplotlib.pyplot as plt

g = ggplot(diamonds, aes(x='price', color='cut')) + \
    geom_density()

g.draw()

plt.title("Density ggplot-based plot in Bokeh.")

pyplot.show_bokeh(plt.gcf(), filename="density.html")
````
