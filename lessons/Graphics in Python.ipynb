## Matplotlib

This is probably the most widely used plotting toolkit. It is what is loaded when we start a notebook with the `--pylab=inline` argument. Matplotllib has two interfaces:
* a functional API that mimics Matlab (pylab)
* an object-oriented version (matplotlib.pyplot, typically aliased to plt)

### Line graph

```python
xs = linspace(0, 4*pi)
pylab.plot(xs, sin(xs), 'r-')
pylab.title('Sine function between $\pi$ and $4\pi$', fontsize=16)
pylab.xticks(linspace(0, 4*pi, 4), ['$0$, $\pi$', '$2\pi$', '$3\pi$', '$4\pi$'], fontsize=16);
```

### Contour plot

```python
x = linspace(-3.0, 3.0, 100) 
y = linspace(-3.0, 3.0, 100) 
X, Y = meshgrid(x, y) 
Z = exp(-(X**2 + Y**2))
plt.contour(X, Y, Z)
plt.contourf(X, Y, Z, cmap=cm.coolwarm)
plt.colorbar();
```

### 3D plots

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100);
```

## Seaborn

This new library is designed for statistical graphics.

### Plot combining KDE with rug/histogram

```python
random.seed(23)
xs = random.normal(0,1,100)

figure(figsize=(8,4))
sns.set_palette("hls")
plt.subplot(1,2,1)
sns.distplot(xs, hist=False, rug=True);
sns.set_palette("coolwarm")
plt.subplot(1,2,2)
sns.distplot(xs, hist=True);
```

### Box and violin plots

```python
df = sm.datasets.get_rdataset('sleep').data
subplot(1,2,1)
sns.boxplot(df.extra, df.group, color="pastel")
subplot(1,2,2)
sns.violinplot(df.extra, df.group, color="pastel");
```

### 2D density plots

```python
random.seed(23)
xs = np.random.multivariate_normal([0,0], eye(2), 100)
data = pd.DataFrame(xs, columns=("X", "Y"))
sns.jointplot("X", "Y", data, kind="kde", color='orange');
```

### Linear model plots
