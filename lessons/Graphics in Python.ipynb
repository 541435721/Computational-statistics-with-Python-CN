### Matplotlib

This is probably the most widely used plotting toolkit. It is what is loaded when we start a notebook with the `--pylab=inline` argument. Matplotllib has two interfaces:
* a functional API that mimics Matlab (pylab)
* an object-oriented version (matplotlib.pyplot, typically aliased to plt)

#### Using the pylab interface

```python
# Line graph

xs = linspace(0, 4*pi)
pylab.plot(xs, sin(xs), 'r-')
pylab.title('Sine function between $\pi$ and $4\pi$', fontsize=16)
pylab.xticks(linspace(0, 4*pi, 4), ['$0$, $\pi$', '$2\pi$', '$3\pi$', '$4\pi$'], fontsize=16);
```

```python
# Bar chart

df = sm.datasets.longley.load_pandas().data
plt.bar(df['YEAR'], df['UNEMP'], align='center', alpha=0.5)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Number of unemployed ', fontsize=16)
plt.xlim([df['YEAR'].min()-1, df['YEAR'].max()+1])
```

```python
# Contour plot

xlist = linspace(-3.0, 3.0, 100) 
ylist = linspace(-3.0, 3.0, 100) 
X, Y = meshgrid(xlist, ylist) 
Z = sqrt(X**2 + Y**2) 
plt.contourf(X, Y, Z, cmap=cm.summer)
plt.colorbar();
```
