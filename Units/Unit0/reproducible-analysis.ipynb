Setup
----

Suppose we have some gene expression data from a case control experiment in a CSV files that reside in a folder "CoolExperiments". Each folder contains files that have the format cases.csv or ctrls.csv. Each CSV file contains a header with $n$ gene names, along with $n$ rows of gene expression levels, with each row represening a speicifc case or control subject.

We want to read in the data, perform pairwise t-tests to find the genes that difer most between cases and controls, plot the heatmap and reprot the top 10 most different genes in a reprot. We would also like to do this in a reproducible way, since new data rows may be added as additional subjects are recruited into the study, or we might want to find the top 5 differnt genes only.

Note that all data is fake and generated from random normal variates, so any pattern we see is a nicce illustration of noise discovery.

```python
import os
import sys
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
%precision 4
```

```python
from itertools import permutations, islice
```

```python
folder = 'lecture-01'
n = 25
p = 20000

names = list(islice((''.join(perm) for perm in permutations('abcdefghi')), 0, p))
cases = pd.DataFrame(np.random.normal(0, 1, (n, p)) , columns=names)
ctrls = pd.DataFrame(np.random.normal(0, 1, (n, p)), columns=names)
cases.to_csv(os.path.join(folder, 'cases.csv'), index=False)
ctrls.to_csv(os.path.join(folder, 'ctrls.csv'), index=False)
```

```python
import pandas as pd
import scipy.stats as st

cases = pd.read_csv(os.path.join(folder, 'cases.csv'))
ctrls = pd.read_csv(os.path.join(folder, 'ctrls.csv'))
```

```python
ntop = 50

genes = ctrls.columns
scores, pvals = st.ttest_ind(cases, ctrls)
top_idx = np.argsort(pvals)[:ntop]
df = pd.DataFrame(np.array([genes[top_idx], pvals[top_idx]]).T, columns=['Gene', 'p'])
```

```python
from tabulate import tabulate 

print tabulate(df, headers=list(df.columns), tablefmt="latex") 
```

```python
import seaborn as sns

cmap = sns.diverging_palette(255, 1, n=3, as_cmap=True)
sns.clustermap(pd.concat([ctrls.ix[:, top_idx].sort(axis=1), cases.ix[:, top_idx].sort(axis=1)]).T, cmap=cmap);
plt.savefig(os.path.join(folder, 'heatmap.png'))
```
